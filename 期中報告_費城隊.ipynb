# Grid Search Cross-Validation

```python
from sklearn.feature_selection import RFE
from sklearn.model_selection import GridSearchCV

# creating a KFold object with 5 splits 
folds = KFold(n_splits = 5, shuffle = True, random_state = 100)

# specify range of hyperparameters
hyper_params = [{'n_features_to_select': list(range(1, 14))}]

# specify model
lm = LinearRegression()
lm.fit(X_std, y_train)
rfe = RFE(lm)             

# set up GridSearchCV() # scoring= r2 / 'neg_mean_squared_error'/'explained_variance'
model_cv = GridSearchCV(estimator = rfe, 
                        param_grid = hyper_params, 
                        scoring= 'r2', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True,
                        refit=True)      
# fit the model
model_cv.fit(X_std, y_train) 
```

```python
means = model_cv.cv_results_['mean_test_score']
params = model_cv.cv_results_['params']
print(means,params)
print(model_cv.best_params_)
```

```python
# cv results
cv_results = pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
# plotting cv results
plt.figure(figsize=(16,6))

plt.plot(cv_results["param_n_features_to_select"], cv_results["mean_test_score"])
plt.plot(cv_results["param_n_features_to_select"], cv_results["mean_train_score"])
plt.xlabel('number of features')
plt.ylabel('r-squared')
plt.axvline(x=5, color='r', linestyle='dashed', linewidth=3)
plt.title("Optimal Number of Features")
plt.legend(['mean of test fold score', 'mean of train fold score'], loc='upper left')
```

# Final muti-linear regression model

```python
# final muti-linear regression model
from sklearn.preprocessing import PolynomialFeatures
quadratic = PolynomialFeatures(degree=2)


cols=['LSTAT', 'RM', 'PTRATIO','DIS', 'NOX']
p = len(cols)

new_X_train = X_train[cols]
new_X_train['LSTAT_square'] = quadratic.fit_transform(new_X_train[['LSTAT']])[:,2]
new_X_test = X_test[cols]
new_X_test['LSTAT_square']  = quadratic.fit_transform(new_X_test[['LSTAT']])[:,2]
lm = LinearRegression()

# fit the model 
lm.fit(new_X_train, y_train) #建立 model
print('intercept =', lm.intercept_ )
print('Coef =', lm.coef_)

# Find the R^2
r2_train = lm.score(new_X_train, y_train)
r2_test = lm.score(new_X_test, y_test)
print('The R-square for train is : ', r2_train )
print('The R-square for test is : ', r2_test )

r2_adj = 1 - ((1 - r2_test) * ((y_test.shape[0] - 1)/(y_test.shape[0] - p -1)) )
print('The Adjusted R^2 for test is : ' + str(r2_adj))

#預測
Y_train_hat = lm.predict(new_X_train) 
Y_hat = lm.predict(new_X_test) 
from sklearn.metrics import mean_squared_error
print('The mean square error of price and predicted value using multifit is : ', \
      mean_squared_error(y_test, Y_hat))
```
